```python
"""
---------------------------------------------
Data Information
: Temperature data form SUWON from 1964.01.01 to 2021.04.25
: Temperature data form GANGNEUNG from 1964.01.01 to 2021.04.25
---------------------------------------------
Column Information
: 날짜 / 지점 / 평균기온(℃) / 최저기온(℃) / 최고기온(℃)
---------------------------------------------
Encoding type
: KOR-ECU (cp949)
---------------------------------------------
Redeaing method
: pandas package
---------------------------------------------
Backup position
: suwon_data / gangneung_data
---------------------------------------------
"""
# Read csv file with Built-in function
f = open('/content/drive/MyDrive/Data_ML_Midterm/Statistic_Resources/suwon_temperature_from_19640101_to_20210425.csv', 'r', encoding = 'cp949')

f1 = open('/content/drive/MyDrive/Data_ML_Midterm/Statistic_Resources/gangneung_temperature_from_19640101_20210425.csv', 'r', encoding = 'cp949')


# read csv file with pandas package
suwon_data = pd.read_csv(f)
gangneung_data = pd.read_csv(f1)


"""
---------------------------------------------
Array Declaration
: suwon_temperature_array           -> copy original data
: suwon_average_temperature_array   -> information of "평균기온(℃)"

: gangneung_temperature_array           -> copy original data
: gangneung_average_temperature_array   -> information of "평균기온(℃)"
---------------------------------------------
"""
# Array Declaration
suwon_temperature_array = []
gangneung_temperature_array = []

suwon_average_temperature_array = []
gangneung_average_temperature_array = []


# copy original data(suwon_data)
suwon_temperature_array = suwon_data.copy()
gangneung_temperature_array = gangneung_data.copy()


"""
---------------------------------------------
Data Double check
: .head() -> check data format
: .info() -> check data type and null-data
---------------------------------------------
"""

### SUWON
print('\n------------------------------------------------------------')
print("## 수원")
print('------------------------------------------------------------\n')

# check data format
print(suwon_data.head())

print('\n------------------------------------------------------------\n')

# check data type and null-data
print(suwon_data.info())

print('\n------------------------------------------------------------\n')




### GANGNEUNG
print('\n------------------------------------------------------------')
print("## 강릉")
print("------------------------------------------------------------\n")


# check data format
print(gangneung_data.head())

print('\n------------------------------------------------------------\n')

# check data type and null-data
print(gangneung_data.info())

print('\n------------------------------------------------------------\n')
```

##<h1><strong>Synthesis of Data </strong><h1>

> <h2>수원, 강릉 날씨 종합적인 비교</h2>

<br>

- Factor1 : 온도 그래프 겹쳐서 비교
- Factor2 : 매년 온도 차이 정도 (MAX 평균온도 - MIN 평균온도) &nbsp; 비교

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; 온도 차이 큰경우 &nbsp; &nbsp; &nbsp;-> Max(30도) - Min(-20도) == 온도 차이 정도(50)

&nbsp; &nbsp; &nbsp; &nbsp; &nbsp; 온도 차이 작은경우 &nbsp;-> Max(30도) - Min(-10도) == 온도 차이 정도(40)

<br>

```python

"""
---------------------------------------------
Array Declaration
: suwon_avg_temperature_difference_data       -> value of ( highest avg temperature + lowest avg temperature )
: gangneung_avg_temperature_difference_data   -> value of ( highest avg temperature + lowest avg temperature )
---------------------------------------------
variable Declaration
: max_temp            -> max temperature from SUWON
: min_temp            -> min temperature from SUWON

: max_temp        -> max temperature from GANGNEUNG
: min_temp        -> min temperature from GANGNEUNG

: year_count          -> year count value to seperate every year
: year_count      -> year count value to seperate every year
---------------------------------------------
"""

# Array Declaration
suwon_avg_temperature_difference_data = []
gangneung_avg_temperature_difference_data = []


# variable Declaration
max_temp = -999
min_temp = 999

year_count = 1964




"""
###################################################################
SUWON 
---------------------------------------------
Iterating options
: iterate indx within suwon_data(csv file)
---------------------------------------------
Iterating procedure

: 1st [if] -> compare value[year] in csv value[year_count] 
             -> csv Bigger
              - value[year_count] count up
              - append value[max - min] to  "suwon_avg_temperature_difference_data"
              - initialization max value
              - initialization min value

: 2nd [if] -> compare [previous max value] with [new value]
             -> [new value] is Bigger
              - update max value 

: 3rd [if] -> compare [previous min value] with [new value]
             -> [new value] is Smaller
              - update min value 

---------------------------------------------
"""
# for loop
for indx in suwon_temperature_array.index:

  ## 1st [if]
  # 날짜(year - month - data)  ->  [year] - [year_count] 
  if (float(suwon_temperature_array["날짜"][indx].split('-')[0]) - year_count) == 1:
  
    # count up [year_count]
    year_count += 1

    # append [max_temp - min_temp] to "suwon_avg_temperature_difference_data"
    suwon_avg_temperature_difference_data.append(max_temp - min_temp)

    # initialization [max and min] value
    max_temp = -999
    min_temp = 999



  ## 2nd [if]
  # find the biggest value
  if (suwon_temperature_array["평균기온(℃)"][indx] > max_temp):
    max_temp = suwon_temperature_array["평균기온(℃)"][indx]

  ## 3rd [if]
  # find the smallest value
  if (suwon_temperature_array["평균기온(℃)"][indx] < min_temp):
    min_temp = suwon_temperature_array["평균기온(℃)"][indx]
    



"""
###################################################################
GANGNEUNG
---------------------------------------------
Iterating options
: iterate indx within gangneung_data(csv file)
---------------------------------------------
Iterating procedure

: 1st [if] -> compare value[year] in csv value[year_count] 
             -> csv Bigger
              - value[year_count] count up
              - append value[max - min] to  "gangneung_avg_temperature_difference_data"
              - initialization max value
              - initialization min value

: 2nd [if] -> compare [previous max value] with [new value]
             -> [new value] is Bigger
              - update max value 

: 3rd [if] -> compare [previous min value] with [new value]
             -> [new value] is Smaller
              - update min value 
              
---------------------------------------------
"""

# initialization
max_temp = -999
min_temp = 999
year_count = 1964


# for loop
for indx in gangneung_temperature_array.index:

  ## 1st [if]
  # 날짜(year - month - data)  ->  [year] - [year_count] 
  if (float(gangneung_temperature_array["날짜"][indx].split('-')[0]) - year_count) == 1:

    # count up [year_count]
    year_count += 1

    # append [max_temp - min_temp] to "suwon_avg_temperature_difference_data"
    gangneung_avg_temperature_difference_data.append(max_temp - min_temp)

    # initialization [max and min] value
    max_temp = -999
    min_temp = 999


  ## 2nd [if]
  # find the biggest value
  if (gangneung_temperature_array["평균기온(℃)"][indx] > max_temp):
    max_temp = gangneung_temperature_array["평균기온(℃)"][indx]

  ## 3rd [if]
  # find the smallest value
  if (gangneung_temperature_array["평균기온(℃)"][indx] < min_temp):
    min_temp = gangneung_temperature_array["평균기온(℃)"][indx]






"""
---------------------------------------------
Ploting Options

: total figure size     => 50 by 10

####################################################################
######## Subplot [First] (2,1,1)

: Title               => "수원 평균온도 from 1964.0101 to 2021.04.25"

: (1) plotting type   => default
                      => color: c, markers: point, label: "수원"
: (2) plotting type   => default
                      => color: m, markers: point, label: "강릉" 

: xlabel              => 온도 samples
: ylabel              => 섭씨
: font_collison       => unicode_minus = False
: grid                => ON
: legend              => loc = 2, fontsize = large
####################################################################


####################################################################
######## Subplot [Second] (2,1,2)

: Title               => "강릉 평균온도 from 1964.0101 to 2021.04.25"

: (1) plotting type   => default
                      => color: c, markers: D-, label: "수원 연중 기온편차" 
: (2) plotting type   => default
                      => color: m, markers: D-, label: "강릉 연중 기온편차"                    

: xlabel              => 매년 온도차이 samples
: ylabel              => 섭씨


: position_table      => Arrow position table
: annotate            => show [arrow] and [text] on Graph
                      => xy: [arrow] position, [xytext]: text position
                      => ha: arrow start point(left, center, right)
                      => arrowprops: arrow's head and body option


: grid                => ON
: legend              => loc = 2, fontsize = large
####################################################################

: subplot_adjust  => hspace = 0.5
---------------------------------------------
"""
## Total option
plt.figure(figsize=(50,10))

####################################################################
# First Subplot options
plt.subplot(2,1,1)
plt.title("수원과 강릉의 평균기온 분포도 1964.01.01 ~ 2021.04.25")
plt.plot(suwon_average_temperature_array, 'c.', label= "수원", markersize=5)
plt.plot(gangneung_average_temperature_array,'m.', label= "강릉", markersize=5)
plt.xlabel("온도 samples")
plt.ylabel("섭씨")
plt.rc('axes', unicode_minus=False)
plt.legend(loc=2, fontsize='large')
####################################################################



####################################################################
# Second Subplot options
plt.subplot(2,1,2)
plt.title("수원과 강릉의 매년 최고평균기온과 최저평균기온의 합 ")
plt.plot(suwon_avg_temperature_difference_data, 'cD-', label="수원 연중 기온편차")
plt.plot(gangneung_avg_temperature_difference_data, 'mD-', label="강릉 연중 기온편차")
plt.xlabel("매년 온도차이 samples")
plt.ylabel("섭씨")


position_table = [
    
    [4, 41.4], [13,42.1], [28,36.4], [36,37.6], [38,39.9], [43,34.9]

]

for x_pos, y_pos in position_table:

  plt.annotate('강릉  >  수원', xy=(x_pos, y_pos+0.5), xytext=(x_pos,y_pos+3),
            fontsize=11, ha='center',
            arrowprops=dict(facecolor='black', width=1, shrink=0.1, headwidth=10))

plt.text(0.3, 34.5, "숫자가 클수록 온도 차이가 크다", fontsize=15)
plt.grid()
plt.legend(loc=2, fontsize='large')
####################################################################



## Total option
plt.subplots_adjust(hspace=0.5)
### ploting graph
plt.show
```
