```python
"""
---------------------------------------------
Data Information
: Temperature data form SUWON from 1964.01.01 to 2021.04.25
---------------------------------------------
Column Information
: 날짜 / 지점 / 평균기온(℃) / 최저기온(℃) / 최고기온(℃)
---------------------------------------------
Encoding type
: KOR-ECU (cp949)
---------------------------------------------
Redeaing method
: pandas package
---------------------------------------------
Backup position
: suwon_data
---------------------------------------------
"""
# Read csv file with Built-in function
f = open('/content/drive/MyDrive/Data_ML_Midterm/Statistic_Resources/suwon_temperature_from_19640101_to_20210425.csv', 'r', encoding = 'cp949')

# read csv file with pandas package
suwon_data = pd.read_csv(f)


"""
---------------------------------------------
Array Declaration
: suwon_temperature_array           -> copy original data
: suwon_average_temperature_array   -> information of "평균기온(℃)"
---------------------------------------------
"""
# Array Declaration
suwon_temperature_array = []
suwon_average_temperature_array = []

# copy original data(suwon_data)
suwon_temperature_array = suwon_data.copy()


"""
---------------------------------------------
Data Double check
: .head() -> check data format
: .info() -> check data type and null-data
---------------------------------------------
"""
print('\n------------------------------------------------------------\n')

# check data format
print(suwon_data.head())

print('\n------------------------------------------------------------\n')

# check data type and null-data
print(suwon_data.info())

print('\n------------------------------------------------------------\n')
```

```python

"""
---------------------------------------------
Iterating options
: iterate indx within suwon_data(csv file)
---------------------------------------------
Iterating procedure

: 1st -> main loop == csv file loop with [index number]

: 2nd -> append "평균기온(℃)" data to "suwon_average_temperature_array"
         data[coulmn_name][index_number]

---------------------------------------------
"""

for indx in suwon_temperature_array.index:
  suwon_average_temperature_array.append(suwon_temperature_array["평균기온(℃)"][indx])




"""
---------------------------------------------
Ploting Options
: figure size     => 40 by 5
: Title           => "수원 평균온도 from 1964.0101 to 2021.04.25"

: plotting type   => default
                  => color: m, markers: point  

: xlabel          => 섭씨
: ylabel          => samples
: font_collison   => unicode_minus = False
: grid            => ON
: legend          => loc = 2, fontsize = large
---------------------------------------------
"""

## plot options
plt.figure(figsize=(40,5))
plt.title("수원 평균온도 from 1964.0101 to 2021.04.25")
plt.plot(suwon_average_temperature_array, 'm.', label="수원 평균기온")
plt.xlabel("samples")
plt.ylabel("섭씨")
plt.rc('axes', unicode_minus=False)
plt.grid()
plt.legend(loc=2, fontsize='large')

### ploting graph
plt.show

```

##<h1><strong>Conculusion of Analyze </strong><h1>

> <h2>수원, 강릉 날씨 종합적인 비교</h2>

<br>

- Factor1 : 분포도 비교
- Factor2 : 평균온도 data중 최고온도와 최저온도를 더한 값의 매년 변화

<br>

```python

"""
---------------------------------------------
Array Declaration
: suwon_avg_temperature_difference_data       -> value of ( highest avg temperature + lowest avg temperature )
: gangneung_avg_temperature_difference_data   -> value of ( highest avg temperature + lowest avg temperature )
---------------------------------------------
variable Declaration
: suwon_max_temp            -> max temperature from SUWON
: suwon_min_temp            -> min temperature from SUWON

: gangneung_max_temp        -> max temperature from GANGNEUNG
: gangneung_min_temp        -> min temperature from GANGNEUNG

: suwon_year_count          -> year count value to seperate every year
: gangneung_year_count      -> year count value to seperate every year
---------------------------------------------
"""

# Array Declaration
suwon_avg_temperature_difference_data = []
gangneung_avg_temperature_difference_data = []


# variable Declaration
suwon_max_temp = -999
suwon_min_temp = 999

gangneung_max_temp = -999
gangneung_min_temp = 999

suwon_year_count = 1964
gangneung_year_count = 1964



"""
###################################################################
SUWON 
---------------------------------------------
Iterating options
: iterate indx within suwon_data(csv file)
---------------------------------------------
Iterating procedure

: 1st [if] -> compare value[year] in csv value[year_count] 
             -> csv Bigger
              - value[year_count] count up
              - append value[max + min] to  "suwon_avg_temperature_difference_data"
              - initialization max value
              - initialization min value

: 2nd [if] -> compare [previous max value] with [new value]
             -> [new value] is Bigger
              - update max value 

: 3rd [if] -> compare [previous min value] with [new value]
             -> [new value] is Smaller
              - update min value 

---------------------------------------------
"""
# for loop
for indx in suwon_temperature_array.index:

  ## 1st [if]
  # 날짜(year - month - data)  ->  [year] - [suwon_year_count] 
  if (float(suwon_temperature_array["날짜"][indx].split('-')[0]) - suwon_year_count) == 1:
    
    # count up [suwon_year_count]
    suwon_year_count += 1

    # append [suwon_max_temp + suwon_min_temp] to "suwon_avg_temperature_difference_data"
    suwon_avg_temperature_difference_data.append(suwon_max_temp + suwon_min_temp)

    # initialization [max and min] value
    suwon_max_temp = -999
    suwon_min_temp = 999



  ## 2nd [if]
  # find the biggest value
  if (suwon_temperature_array["평균기온(℃)"][indx] > suwon_max_temp):
    suwon_max_temp = suwon_temperature_array["평균기온(℃)"][indx]

  ## 3rd [if]
  # find the smallest value
  if (suwon_temperature_array["평균기온(℃)"][indx] < suwon_min_temp):
    suwon_min_temp = suwon_temperature_array["평균기온(℃)"][indx]
    



"""
###################################################################
GANGNEUNG
---------------------------------------------
Iterating options
: iterate indx within gangneung_data(csv file)
---------------------------------------------
Iterating procedure

: 1st [if] -> compare value[year] in csv value[year_count] 
             -> csv Bigger
              - value[year_count] count up
              - append value[max + min] to  "gangneung_avg_temperature_difference_data"
              - initialization max value
              - initialization min value

: 2nd [if] -> compare [previous max value] with [new value]
             -> [new value] is Bigger
              - update max value 

: 3rd [if] -> compare [previous min value] with [new value]
             -> [new value] is Smaller
              - update min value 
              
---------------------------------------------
"""


# for loop
for indx in gangneung_temperature_array.index:

  ## 1st [if]
  # 날짜(year - month - data)  ->  [year] - [suwon_year_count] 
  if (float(gangneung_temperature_array["날짜"][indx].split('-')[0]) - gangneung_year_count) == 1:
    
    # count up [suwon_year_count]
    gangneung_year_count += 1

    # append [suwon_max_temp + suwon_min_temp] to "suwon_avg_temperature_difference_data"
    gangneung_avg_temperature_difference_data.append(gangneung_max_temp + gangneung_min_temp)

    # initialization [max and min] value
    gangneung_max_temp = -999
    gangneung_min_temp = 999


  ## 2nd [if]
  # find the biggest value
  if (gangneung_temperature_array["평균기온(℃)"][indx] > gangneung_max_temp):
    gangneung_max_temp = gangneung_temperature_array["평균기온(℃)"][indx]

  ## 3rd [if]
  # find the smallest value
  if (gangneung_temperature_array["평균기온(℃)"][indx] < gangneung_min_temp):
    gangneung_min_temp = gangneung_temperature_array["평균기온(℃)"][indx]






"""
---------------------------------------------
Ploting Options

: total figure size     => 50 by 10

######## Subplot [First] (2,1,1)

: Title               => "수원 평균온도 from 1964.0101 to 2021.04.25"

: (1) plotting type   => default
                      => color: c, markers: point, label: "수원"
: (2) plotting type   => default
                      => color: m, markers: point, label: "강릉" 

: xlabel              => 온도 samples
: ylabel              => 섭씨
: font_collison       => unicode_minus = False
: grid                => ON
: legend              => loc = 2, fontsize = large



######## Subplot [Second] (2,1,2)

: Title               => "수원 평균온도 from 1964.0101 to 2021.04.25"

: (1) plotting type   => default
                      => color: c, markers: point, label: "수원 기온편차" 
: (2) plotting type   => default
                      => color: c, markers: point, label: "강릉 기온편차"                    

: xlabel              => 매년 온도차이 samples
: ylabel              => 섭씨
: grid                => ON
: legend              => loc = 2, fontsize = large


: subplot_adjust  => hspace = 0.5
---------------------------------------------
"""
## Total option
plt.figure(figsize=(50,10))


# First Subplot options
plt.subplot(2,1,1)
plt.title("수원과 강릉의 평균기온 분포도 1964.01.01 ~ 2021.04.25")
plt.plot(suwon_average_temperature_array, 'c.', label= "수원", markersize=5)
plt.plot(gangneung_average_temperature_array,'m.', label= "강릉", markersize=5)
plt.xlabel("온도 samples")
plt.ylabel("섭씨")
plt.rc('axes', unicode_minus=False)
plt.grid()
plt.legend(loc=2, fontsize='large')


# Second Subplot options
plt.subplot(2,1,2)
plt.title("수원과 강릉의 매년 최고평균기온과 최저평균기온의 합 ")
plt.xlabel("매년 온도차이 samples")
plt.ylabel("섭씨")
plt.plot(suwon_avg_temperature_difference_data, 'c-', label="수원 기온편차")
plt.plot(gangneung_avg_temperature_difference_data, 'm-', label="강릉 기온편차")
plt.grid()
plt.legend(loc=2, fontsize='large')

## Total option
plt.subplots_adjust(hspace=0.5)

### ploting graph
plt.show

```
